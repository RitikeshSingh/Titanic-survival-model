```python
training_set.drop(['Cabin','Name','Ticket','Embarked'],axis = 1,inplace = True)
```

```python
def agefill(data):
    age = data[0]
    sex = data[1]
    
    if pd.isnull(age):
        if sex is 'male':
            return 29
        else:
            return 25
    else:
        return age
```

```python
training_set['Age'] = training_set[['Age','Sex'] ].apply(agefill,axis = 1)
```

```python
sns.heatmap(training_set.isnull(),yticklabels = False,cbar = False,cmap = 'Blues')
```

```python
training_set['Age'].hist(bins = 20)
```

```python
training_set.drop(['PassengerId'],axis = 1,inplace =True)
```

```python
training_set
```

```python
male = pd.get_dummies(training_set['Sex'],drop_first = True)
```

```python
male
```

```python
training_set.drop(['Sex'],axis = 1,inplace = True)
```

```python
training_set
```

```python
training_set=pd.concat([training_set,male],axis = 1)
```

```python
training_set
```

```python
x=training_set.drop('Survived',axis = 1).values
```

```python
x
```

```python
y=training_set['Survived'].values
```

```python
y
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train ,y_test = train_test_split(x,y,test_size = 0.2, random_state = 20)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train,y_train)
```

```python
y_predict = classifier.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)
sns.heatmap(cm,annot = True,fmt = 'd')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```
